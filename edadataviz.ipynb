### TASK 2: Visualize the relationship between Payload Mass and Launch Site

### TASK  3: Visualize the relationship between success rate of each orbit type


Next, we want to visually check if there are any relationship between success rate and orbit type.


Let's create a `bar chart` for the sucess rate of each orbit


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate success rate for each orbit
success_rate = df.groupby('Orbit')['Class'].mean().reset_index()
success_rate.columns = ['Orbit', 'SuccessRate']

# Sort by success rate in descending order
success_rate = success_rate.sort_values('SuccessRate', ascending=False)

# Create the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x='Orbit', y='SuccessRate', data=success_rate)

# Customize the plot
plt.title('Success Rate by Orbit Type', fontsize=16)
plt.xlabel('Orbit Type', fontsize=12)
plt.ylabel('Success Rate', fontsize=12)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

# Add value labels on top of each bar
for i, v in enumerate(success_rate['SuccessRate']):
    plt.text(i, v, f'{v:.2%}', ha='center', va='bottom')

# Adjust y-axis to show percentages
plt.ylim(0, 1)
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: '{:.0%}'.format(y)))

plt.tight_layout()
plt.show()
```

Analyze the plotted bar chart to identify which orbits have the highest success rates.

### TASK  6: Visualize the launch success yearly trend


You can plot a line chart with x axis to be <code>Year</code> and y axis to be average success rate, to get the average launch success trend.


The function will help you get the year from the date:


```python
# A function to Extract years from the date 
year=[]
def Extract_year():
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year
Extract_year()
df['Date'] = year
df.head()
    
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your DataFrame with the full dataset
# If not, you'll need to load your data into df first

# Extract year from the Date column
df['Year'] = pd.to_datetime(df['Date']).dt.year

# Calculate success rate for each year
success_rate = df.groupby('Year')['Class'].mean().reset_index()
success_rate.columns = ['Year', 'SuccessRate']

# Create the line plot
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='SuccessRate', data=success_rate, marker='o')

# Customize the plot
plt.title('Launch Success Rate by Year', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Success Rate', fontsize=12)

# Adjust y-axis to show percentages
plt.ylim(0, 1)
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: '{:.0%}'.format(y)))

# Add value labels on each point
for x, y in zip(success_rate['Year'], success_rate['SuccessRate']):
    plt.text(x, y, f'{y:.2%}', ha='center', va='bottom')

# Ensure x-axis shows all years
plt.xticks(success_rate['Year'])

plt.tight_layout()
plt.show()
```

you can observe that the sucess rate since 2013 kept increasing till 2020

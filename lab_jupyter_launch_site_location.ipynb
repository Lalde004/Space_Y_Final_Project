```python
import folium
import pandas as pd
from folium.plugins import MarkerCluster, MousePosition
from folium.features import DivIcon

data = [
    [1, 6104.959412, 'LEO', 'CCAFS SLC 40', 1, False, False, False, 'NaN', 1.0, 0, 'B0003', -80.577366, 28.561857],
    [2, 525.000000, 'LEO', 'CCAFS SLC 40', 1, False, False, False, 'NaN', 1.0, 0, 'B0005', -80.577366, 28.561857],
    [3, 677.000000, 'ISS', 'CCAFS SLC 40', 1, False, False, False, 'NaN', 1.0, 0, 'B0007', -80.577366, 28.561857],
    [4, 500.000000, 'PO', 'VAFB SLC 4E', 1, False, False, False, 'NaN', 1.0, 0, 'B1003', -120.610829, 34.632093],
    [5, 3170.000000, 'GTO', 'CCAFS SLC 40', 1, False, False, False, 'NaN', 1.0, 0, 'B1004', -80.577366, 28.561857]
]

columns = ['FlightNumber', 'PayloadMass', 'Orbit', 'LaunchSite', 'Flights', 'GridFins', 'Reused', 'Legs', 'LandingPad', 'Block', 'ReusedCount', 'Serial', 'Longitude', 'Latitude']

df = pd.DataFrame(data, columns=columns)

# Create a map
launch_sites_map = folium.Map(location=[30, -90], zoom_start=4)

# Add MousePosition plugin to the map
formatter = "function(num) {return L.Util.formatNum(num, 3) + ' º ';};"
mouse_position = MousePosition(
    position='topright',
    separator=' | ',
    empty_string='NaN',
    lng_first=True,
    num_digits=20,
    prefix='Coordinates:',
    lat_formatter=formatter,
    lng_formatter=formatter
)
launch_sites_map.add_child(mouse_position)

# Create a MarkerCluster
marker_cluster = MarkerCluster().add_to(launch_sites_map)

# Add markers for each launch site
for idx, row in df.drop_duplicates(subset=['LaunchSite']).iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=row['LaunchSite'],
        icon=folium.Icon(color='red', icon='info-sign')
    ).add_to(marker_cluster)

# Display the map
launch_sites_map
```

```python
import folium
from folium.plugins import MousePosition, MarkerCluster
from folium.features import DivIcon

# NASA Johnson Space Center's coordinates
nasa_coordinate = [29.559684888503615, -95.0830971930759]

# Create a map centered on the NASA Johnson Space Center
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Add MousePosition plugin to the map
formatter = "function(num) {return L.Util.formatNum(num, 5) + ' º ';};"
mouse_position = MousePosition(
    position='topright',
    separator=' | ',
    empty_string='NaN',
    lng_first=True,
    num_digits=20,
    prefix='Coordinates:',
    lat_formatter=formatter,
    lng_formatter=formatter
)
site_map.add_child(mouse_position)

# Create MarkerClusters for successful and failed launches
success_cluster = MarkerCluster(name="Successful Launches").add_to(site_map)
failed_cluster = MarkerCluster(name="Failed Launches").add_to(site_map)

# Add markers for each launch
for idx, launch in spacex_df.iterrows():
    coordinate = [launch['Lat'], launch['Long']]
    
    if launch['class'] == 1:  # Successful launch
        folium.Marker(
            coordinate,
            popup=f"Successful launch from {launch['Launch Site']}",
            icon=folium.Icon(color='green', icon='check')
        ).add_to(success_cluster)
    else:  # Failed launch
        folium.Marker(
            coordinate,
            popup=f"Failed launch from {launch['Launch Site']}",
            icon=folium.Icon(color='red', icon='times')
        ).add_to(failed_cluster)

# Add circles for each unique launch site
for idx, site in launch_sites_df.iterrows():
    coordinate = [site['Lat'], site['Long']]
    
    # Count successful and failed launches for this site
    site_launches = spacex_df[spacex_df['Launch Site'] == site['Launch Site']]
    success_count = site_launches['class'].sum()
    failed_count = len(site_launches) - success_count
    
    # Add Circle
    folium.Circle(
        coordinate, 
        radius=1000, 
        color='blue', 
        fill=True,
        popup=f"{site['Launch Site']}<br>Successful: {success_count}<br>Failed: {failed_count}"
    ).add_to(site_map)
    
    # Add Marker
    folium.map.Marker(
        coordinate,
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html=f'<div style="font-size: 12; color:blue;"><b>{site["Launch Site"]}</b></div>',
        )
    ).add_to(site_map)

# Create a blue circle and marker at NASA Johnson Space Center's coordinate
folium.Circle(
    nasa_coordinate, 
    radius=1000, 
    color='#d35400', 
    fill=True
).add_child(folium.Popup('NASA Johnson Space Center')).add_to(site_map)

folium.map.Marker(
    nasa_coordinate,
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>NASA JSC</b></div>',
    )
).add_to(site_map)

# Add layer control
folium.LayerControl().add_to(site_map)

# Adjust the map to show all markers
site_map.fit_bounds(site_map.get_bounds())

# Display the map
site_map
```

```python
import folium
from folium.plugins import MousePosition, MarkerCluster

# Create a map centered on the average of all coordinates
center_lat = spacex_df['Lat'].mean()
center_long = spacex_df['Long'].mean()
site_map = folium.Map(location=[center_lat, center_long], zoom_start=5)

# Add MousePosition plugin to the map
formatter = "function(num) {return L.Util.formatNum(num, 5) + ' º ';};"
mouse_position = MousePosition(
    position='topright',
    separator=' | ',
    empty_string='NaN',
    lng_first=True,
    num_digits=20,
    prefix='Coordinates:',
    lat_formatter=formatter,
    lng_formatter=formatter
)
site_map.add_child(mouse_position)

# Create a MarkerCluster object
marker_cluster = MarkerCluster()

# Add marker_cluster to current site_map
site_map.add_child(marker_cluster)

# Define a function to determine marker color based on launch success
def get_marker_color(class_value):
    return 'green' if class_value == 1 else 'red'

# Iterate through each row in spacex_df
for index, record in spacex_df.iterrows():
    # Create a Marker object with its coordinate
    marker = folium.Marker(
        location=[record['Lat'], record['Long']],
        popup=f"Launch Site: {record['Launch Site']}<br>Class: {'Success' if record['class'] == 1 else 'Failure'}",
        icon=folium.Icon(color='white', icon_color=get_marker_color(record['class']))
    )
    
    # Add the marker to the cluster
    marker_cluster.add_child(marker)

# Add circles for each unique launch site
for idx, site in launch_sites_df.iterrows():
    # Calculate success rate
    site_launches = spacex_df[spacex_df['Launch Site'] == site['Launch Site']]
    success_rate = site_launches['class'].mean()
    
    # Create popup content
    popup_content = f"{site['Launch Site']}<br>Success Rate: {success_rate:.2%}"
    
    # Add Circle
    folium.Circle(
        [site['Lat'], site['Long']], 
        radius=5000,  # Increased radius for visibility
        color=site['marker_color'], 
        fill=True,
        popup=popup_content
    ).add_to(site_map)

# Adjust the map to show all markers
site_map.fit_bounds(site_map.get_bounds())

# Display the map
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from folium.plugins import MousePosition
import pandas as pd
from math import sin, cos, sqrt, atan2, radians

# Function to calculate distance between two points
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance

# Define a set of coastline points (you may want to expand this for more accuracy)
coastline_points = [
    (28.56367, -80.57163),  # Cape Canaveral
    (34.63331, -120.61081),  # Vandenberg
    (28.57325, -80.64689),  # Kennedy Space Center
    # Add more coastline points as needed
]

# Function to find the closest coastline point
def find_closest_coastline(lat, lon):
    closest_point = min(coastline_points, key=lambda point: calculate_distance(lat, lon, point[0], point[1]))
    return closest_point

# Create a map centered on the average of all coordinates
center_lat = launch_sites_df['Lat'].mean()
center_long = launch_sites_df['Long'].mean()
site_map = folium.Map(location=[center_lat, center_long], zoom_start=4)

# Add MousePosition plugin to the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)
site_map.add_child(mouse_position)

# Calculate distances for each launch site
for index, site in launch_sites_df.iterrows():
    site_lat = site['Lat']
    site_long = site['Long']
    
    # Find closest coastline point
    coastline_lat, coastline_lon = find_closest_coastline(site_lat, site_long)
    
    # Calculate distance to coastline
    distance_coastline = calculate_distance(site_lat, site_long, coastline_lat, coastline_lon)
    
    # Add this information to the launch_sites_df
    launch_sites_df.at[index, 'Closest Coastline Lat'] = coastline_lat
    launch_sites_df.at[index, 'Closest Coastline Lon'] = coastline_lon
    launch_sites_df.at[index, 'Distance to Coastline'] = distance_coastline

# Add markers for each launch site
for idx, site in launch_sites_df.iterrows():
    # Create popup content
    popup_content = f"""
    <b>{site['Launch Site']}</b><br>
    Launch Site: Lat: {site['Lat']:.5f}, Lon: {site['Long']:.5f}<br>
    Closest Coastline: Lat: {site['Closest Coastline Lat']:.5f}, Lon: {site['Closest Coastline Lon']:.5f}<br>
    Distance to Coastline: {site['Distance to Coastline']:.2f} km
    """
    
    # Add marker for launch site
    folium.Marker(
        location=[site['Lat'], site['Long']],
        popup=popup_content,
        icon=folium.Icon(color='red', icon='info-sign')
    ).add_to(site_map)
    
    # Add marker for closest coastline point
    folium.Marker(
        location=[site['Closest Coastline Lat'], site['Closest Coastline Lon']],
        popup=f"Closest coastline to {site['Launch Site']}",
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(site_map)
    
    # Add a line connecting the launch site to its closest coastline point
    folium.PolyLine(
        locations=[[site['Lat'], site['Long']], [site['Closest Coastline Lat'], site['Closest Coastline Lon']]],
        color="green",
        weight=2,
        opacity=0.8
    ).add_to(site_map)

# Adjust the map to show all markers
site_map.fit_bounds(site_map.get_bounds())

# Display the map
site_map
```
